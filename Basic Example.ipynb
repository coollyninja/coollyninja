````python
from langchain.llms import HuggingFaceTextGenInference
````

````python
request = """
Create a reactivex pipeline in golang (using `rxgo`) to build a pipeline that performs the following functions:
1. Take in a list of onvif devices matching the following type:
```
type DiscoveredCamera struct {
  Name string `json:"name"`
  ID   string `json:"id"`
  URL  string `json:"url"`
  Hardware string `json:"hardware"`
}
```
2. Pass them in to `rxgo`.
3. For each of the items, query them in parallel using `GetDetails(cam DiscoveredCamera)`, which gathers their profiles.
4. Filter out the devices for which we couldn't gather profiles.
5. Filter out devices which don't have an `RTSP` profile available.
6. Marshal the objects to JSON and publish them to an MQTT topic.

Please focus on the rxgo chain and I will focus on the implementation of the functions to perform the tasks.
"""
````

````python
request = """Please continue"""
````

````python
prompt = f"""###Instruction:
{request}

###Response:
"""
````

````python
llm = HuggingFaceTextGenInference(
    inference_server_url="http://localhost:8080/",
    max_new_tokens=2048 - len(prompt),
    top_k=10,
    top_p=0.95,
    typical_p=0.95,
    temperature=0.01,
    repetition_penalty=1.03,
)
print(llm(prompt))
````
